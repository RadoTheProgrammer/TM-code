```python
import random
```

```python
data = {col:[] for col in ["id","Name","Choice1","Choice2","Choice3"]}
for i in range(40):
    data["id"].append(i)
    data["Name"].append(f"Name_{i}")
    choices = random.sample(range(8),3)
    data["Choice1"].append(choices[0])
    data["Choice2"].append(choices[1])
    data["Choice3"].append(choices[2])
```

```python
df = pd.DataFrame(data)
df.to_csv("input.csv", index=False)
```

```python
from itertools import cycle

serveurs = cycle(["S1", "S2", "S3"])

def attribuer_tache(tache):
    serveur = next(serveurs)
    print(f"{tache} → {serveur}")
```

```python
attribuer_tache("T1")
attribuer_tache("T2")
```

```python
import random

serveurs = ["S1", "S2", "S3"]
poids = [3, 2, 1]

def attribuer_tache(tache):
    serveur = random.choices(serveurs, weights=poids, k=1)[0]
    print(f"{tache} → {serveur}")
attribuer_tache("T1")
attribuer_tache("T2")
attribuer_tache("T3")
attribuer_tache("T4")
attribuer_tache("T5")
attribuer_tache("T6")
```

```python
import heapq

serveurs = [(0, "S1"), (0, "S2"), (0, "S3")]  # (charge, serveur)
heapq.heapify(serveurs)

def attribuer_tache(tache, cout=1):
    charge, serveur = heapq.heappop(serveurs)
    print(f"{tache} → {serveur}")
    print(serveurs)
    heapq.heappush(serveurs, (charge + cout, serveur))
```

```python
attribuer_tache("T1",2)
attribuer_tache("T2")
attribuer_tache("T3")
attribuer_tache("T4")
attribuer_tache("T5")
attribuer_tache("T6")
```
