```python
from itertools import cycle

serveurs = cycle(["S1", "S2", "S3"])

def attribuer_tache(tache):
    serveur = next(serveurs)
    print(f"{tache} → {serveur}")
```

```python
attribuer_tache("T1")
attribuer_tache("T2")
```

```python
import random

serveurs = ["S1", "S2", "S3"]
poids = [3, 2, 1]

def attribuer_tache(tache):
    serveur = random.choices(serveurs, weights=poids, k=1)[0]
    print(f"{tache} → {serveur}")
attribuer_tache("T1")
attribuer_tache("T2")
attribuer_tache("T3")
attribuer_tache("T4")
attribuer_tache("T5")
attribuer_tache("T6")
```

```python
import heapq

serveurs = [(0, "S1"), (0, "S2"), (0, "S3")]  # (charge, serveur)
heapq.heapify(serveurs)

def attribuer_tache(tache, cout=1):
    charge, serveur = heapq.heappop(serveurs)
    print(f"{tache} → {serveur}")
    print(serveurs)
    heapq.heappush(serveurs, (charge + cout, serveur))
```

```python
attribuer_tache("T1",2)
attribuer_tache("T2")
attribuer_tache("T3")
attribuer_tache("T4")
attribuer_tache("T5")
attribuer_tache("T6")
```

```python
import pandas as pd
import mplcursors
import matplotlib.pyplot as plt
df = pd.read_csv(r"Donnees_TMs\Annee_1\results.csv")

plt.scatter(df["Mean"],df["Std"])
# Assuming the scatter plot is created as plt.scatter(df["Mean"], df["Std"])
# Add hover functionality
cursor = mplcursors.cursor(hover=True)

@cursor.connect("add")
def on_add(sel):
    # Get the index of the hovered point
    index = sel.index
    # Display the Id
    sel.annotation.set_text(f'Id: {df.iloc[index]["Id"]}')
```
